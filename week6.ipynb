```python
grades = {
    "Math": 13,
    "Accounts": 12,
    "Programming": 14
}

# 1. Add a new subject
grades["History"] = 8

# 2. Update one grade
grades["Accounts"] = 6

# 3. Print all subjects and grades using .items()
for subject, grade in grades.items():
    print(subject, ":", grade)
```
